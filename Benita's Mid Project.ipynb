# Questions
1. Country with the highest and lowest number of migrants
2. Country with the highest and lowest population.
3. countries whise median age is below 45 and above 45
4. Country with the highest land area.
4. most densed and less densed country

```````````````````````python
#1 Country with the highest number of migrants
world_data1[['Migrants (net)']].max()
```````````````````````

```````````````````````python
#high_mig = world_data1[world_data1['Migrants (net)'] >= 200000]
#high_mig
```````````````````````

```````````````````````python
high_mig = world_data1.sort_values('Migrants (net)', ascending=False).head(8)
high_mig
```````````````````````

```````````````````````python
sns.barplot(high_mig, x='Country/Other', y='Migrants (net)')
plt.xticks(rotation=45)
plt.savefig("figure1.jpg", bbox_inches= 'tight')
plt.show()
```````````````````````

```````````````````````python
world_data1[['Migrants (net)']].min()
```````````````````````

```````````````````````python
low_mig = world_data1.sort_values('Migrants (net)', ascending=False).head()
low_mig
```````````````````````

```````````````````````python
sns.barplot(low_mig, y='Country/Other',x='Migrants (net)', orient='h')
plt.savefig("figure2.jpg", bbox_inches= 'tight')
plt.show()
```````````````````````
